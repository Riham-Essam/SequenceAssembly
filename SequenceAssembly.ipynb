```python
#single read
def get_reads_from_file_single(filename):
    f = open(filename)
    read_len = f.readline().rstrip('\n') #this line contains k-mers used
    #read the file and store it in reads list
    reads =[]
    for line in f:
        if line!= read_len:#Read lines from file as long as line is not equal to k(read_len)
            reads.append(line.rstrip('\n'))
    return read_len, reads


#de bruin graph for single read
def de_bruin_single(reads, len_reads):
    length = int(len_reads)
    #store left k-1mer and right k-1mer in two separate lists
    list_r = []
    list_l = []
    graph = []
    for read in reads:
        # read from begining to k-1 mer
        # from first poistion 1 to end of read
        graph.append((read[:length - 1], read[1:]))
        list_l.append(read[:length - 1])
        list_r.append(read[1:])
    return list_l, list_r, graph

#get start point for single read
def get_startpoint_single(list_l, list_r):
    index_start = 0
#if a left k-1mer found in list_l but not found in list_r this  will be considered as a start point
    for read in list_l:
        index_start += 1
        if read not in list_r:
            return read, index_start - 1
#return start point and its index


#Find EULERIAN path for single read

def find_path_single(list_l, list_r, startpoint, index_start):
    path = []

    path.append(startpoint)
    next_value = list_r[index_start]
    path.append(next_value)
    for i in range(len(list_l)):
        for i in range(len(list_l)):
            if next_value == list_l[i]:
                path.append(list_r[i])
                next_value = list_r[i]

    return path

##reconstruct genome from path
def reconstruct_Genome(path):
    genome = ''
#if the genome is empty the genome will be concatenated with the fist sequence else the genome will be concatenated with the last base of each sequence
    for seq in path:
        if genome == '':
            genome = seq
        else:#last character
            genome += seq[-1]

    return genome


#paired_reads
def get_reads_from_file(filename):
    reads = []
    file = open(filename)
    # split the fisrt line into read_length and gap length
    read_len, gap_len = file.readline().rstrip('\n').split(" ")

    # store every paired-read in the "reads" list
    for line in file:
        if line != read_len:
            reads.append(line.rstrip('\n'))
    return reads, read_len, gap_len


def de_bruin_paired(paired_read_list):
    nodes = []
    # split the read element into prefix and suffix by '|'
    for paired_read in paired_read_list:
        node = paired_read.split('|')
        nodes.append(node)

        # store the prefix column
    left_list = []
    for paired_read in nodes:
        temp_list = []  # to store prefix reads of the same node
        for i in paired_read:
            k_mer = i[:-1]
            temp_list.append(k_mer)
        left_list.append(temp_list)

        # store the suffix column
    right_list = []
    for paired_read in nodes:
        temp_list = []  # to store the suffix reads of the same node
        for i in paired_read:
            k_mer = i[1:]
            temp_list.append(k_mer)
        right_list.append(temp_list)

    return left_list, right_list  # the final data structure is a list of lists





def get_startpoint_paired(right_list, left_list):
    # search about the start node in the suffix list (right_list)
    for i in left_list:
        start_index = left_list.index(i)
        if i not in right_list:
            start_node = i
            return start_node, start_index




def find_end_node(right_list, left_list):
    # search about the end node in the prefix list (left_list)
    for i in right_list:
        if i not in left_list:
            end_node = i
            return end_node



# increase number of recursions
import sys

sys.setrecursionlimit(500000)

path = []


def find_path(index, left_list, right_list, end_node):
    # first condition just to append the start node
    if right_list[index] not in path:
        path.append(right_list[index])

    # Append the secound node by using the same index
    if left_list[index] not in path:
        path.append(left_list[index])

    # Append the nodes between the start and the end node
    if left_list[index] in right_list:
        new_index = right_list.index(left_list[index])
        find_path(new_index, left_list, right_list, end_node)

    # Append the end node
    if end_node not in path:
        path.append(end_node)

    # path=[[prefix, prefix], [suffix,suffix],[prefix,prefix],[suffix,suffix]...]
    # prefix       suffix        prefix         suffix  .....
    # GAC      ->  ACC      ->   CCG        ->  CGA     .....
    # GCG          CGC           GCC            CCG     .....
    return path


def assembly(path):
    prefix_string = ''
    suffix_string = ''

    # lists to store all the prefix k-mers and suffix k-mers
    prefix = []
    suffix = []

    for i in range(len(path)):
        for j in range(len(path[i])):
            if j == 0:
                prefix.append(path[i][j])
            else:
                suffix.append(path[i][j])

    # construct the prefix string by adding every first char from every element from prefix list + the last element
    for i in range(len(prefix)):
        # skip the last element in the list
        if i != len(prefix) - 1:
            prefix_string += prefix[i][0]
        else:
            prefix_string += prefix[i]  # add the last element in the list

    # construct the suffix string by adding every first char from every element from suffix list + the last element
    for i in range(len(suffix)):
        # skip the last element in the list
        if i != len(suffix) - 1:
            suffix_string += suffix[i][0]
        else:
            suffix_string += suffix[i]  # add the whole last element

    whole_genome = ''
    whole_genome += prefix_string
    k = int(read_len)
    d = int(gap_len)
    whole_genome += suffix_string[-1 * (k + d):]
    return whole_genome














#############################################333333
print("Enter File Path : ")
filePath = input(" ")


print("Single or Paired Reads 1 for single , 2 for pair : ")
choice = input(" ")

if choice == "1":
    read_len, reads = get_reads_from_file_single(filePath)
    print("Reads",reads)
    list_l,list_r,Graph = de_bruin_single(reads, read_len)
    print("Graph", Graph)
    print("list left:", list_l)
    print("list right", list_r)
    start_point,index_start=get_startpoint_single(list_l,list_r)
    print("start point", start_point)
    Path = find_path_single(list_l,list_r,start_point,index_start)
    print("Path",Path)
    Genome = reconstruct_Genome(Path)
    print("Reconstruct Genome", Genome)
    print("length of genome", len(Genome))


elif choice == "2":
    reads,read_len,gap_len = get_reads_from_file(filePath)
    print("reads",reads)
    left_list, right_list= de_bruin_paired(reads)
    print("graph:", left_list,right_list)
    start,index=get_startpoint_paired(right_list,left_list)
    print("start:", start)
    end_node=find_end_node(right_list,left_list)
    path = find_path(index,right_list,left_list,end_node)
    print("Path:",path)

    Genome = assembly(path)
    print("Genome:",Genome)
    print("Genome Length",len(Genome))
   
```
